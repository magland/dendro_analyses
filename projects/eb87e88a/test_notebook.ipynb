# Example notebook

This is a simple example notebook that demonstrates how to create a basic plot using matplotlib in Jupyter Notebook.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * t)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t, y)
ax.set_xlabel('time (s)')
ax.set_ylabel('voltage (mV)')
ax.set_title('About as simple as it gets')
plt.show()
```
